# Scraping Data from a Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
# print(soup)
```

```python
table = soup.find_all('table')[1]
soup.find('table', class_ = "wikitable sortable")
```

```python
# print(table)
```

```python
# Run find_all on table, not the Soup itself.
world_titles = table.find_all('th')
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip()for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

## Web Scraping for Medication

```python
url2 = "https://en.wikipedia.org/wiki/List_of_largest_selling_pharmaceutical_products"
page = requests.get(url2)
soup = BeautifulSoup(page.text, 'html')
# print(soup)
```

```python
table = soup.find_all('table')[4]
# print(table)
```

```python
products = table.find_all('th')
products
```

```python
product_table_titles = [title.text.strip() for title in products]
# product_table_titles
```

```python
product_table_titles[8] = 'Patient Expiration Date'
print(product_table_titles)
```

```python
df = pd.DataFrame(columns = product_table_titles)
df
```

```python
column_data = table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_records = [record.text.strip() for record in row_data]
    
    length = len(df)
    df.loc[length] = individual_records
```

```python
df
```

```python
df.to_csv(r'/Users/weezygeezer/documents/workspace/python-programs/Drugs2013.csv', index=False)
```
